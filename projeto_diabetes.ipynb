# Algoritimos de Machine Learning

## Tratamento dos dados

```python
#Separando os dados em variaveis explicativas e variaveis alvo

x = data.drop('diabetes', axis = 1)
y = data['diabetes']
```

```python
#Separando os dados entre treino, validação e teste

from sklearn.model_selection import train_test_split

x, x_teste, y, y_teste = train_test_split(x, y, stratify = y, test_size = 0.05, random_state = 42)
x_treino, x_val, y_treino, y_val = train_test_split(x, y, stratify = y, random_state = 42)
```

```python
#Importando metricas

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

## DesicionTreeClassifier


```python
#Importando o modelo
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(max_depth=3,random_state=42) #instanciando modelo

tree.fit(x_treino,y_treino) #treinando modelo

tree_score_teste = tree.score(x_teste,y_teste) #score do modelo com dados de teste
tree_score_val = tree.score(x_val,y_val) #score do modelo com dados de validação

print(f'A taxa de acerto do modelo DecisionTreeClassifier com os dados de teste é de {tree_score_teste:.2%}')
print(f'A taxa de acerto do modelo DecisionTreeClassifier com os dados de validação é de {tree_score_val:.2%}')
```

```python
#Grafico DecisionTreeClassifier
from sklearn.tree import plot_tree

plt.figure(figsize = (15, 6))
plot_tree(tree, filled = True, class_names = ['nao', 'sim'], fontsize = 5)

plt.show()
```

```python
#Prevendo resultados para x_teste e x_val

tree_predict_teste = tree.predict(x_teste)
tree_predict_val = tree.predict(x_val)
```

```python
#Metricas para o dataset de teste

print(f'Acuracia Teste: {accuracy_score(y_teste,tree_predict_teste):.2%}')
print(f'Precisão Teste: {precision_score(y_teste,tree_predict_teste):.2%}')
print(f'Recall Teste: {recall_score(y_teste,tree_predict_teste):.2%}')
print(f'F1 Score Teste: {f1_score(y_teste,tree_predict_teste):.2%}')
```

```python
#Metricas para o dataset de teste
print(f'Acuracia Validação: {accuracy_score(y_val,tree_predict_val):.2%}')
print(f'Precisão Validação: {precision_score(y_val,tree_predict_val):.2%}')
print(f'Recall Validação: {recall_score(y_val,tree_predict_val):.2%}')
print(f'F1 Score Validação: {f1_score(y_val,tree_predict_val):.2%}')
```

```python
#Matriz de confusão DecisionTreeClassifier para dados de teste

from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

visualizacaotree = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_teste,tree_predict_teste))
visualizacaotree.plot()

plt.show()
```

```python
#Matriz de confusão DecisionTreeClassifier para dados de validação

visualizacaotreeval = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_val,tree_predict_val))
visualizacaotreeval.plot()

plt.show()
```

## KNN(KNeighborsClassifier)

```python
#É necessario a normalização dos dados para execução do modelo

from sklearn.preprocessing import StandardScaler #importando o normalizador
norm = StandardScaler() #instanciando o normalizador

x_treino_normalizado = norm.fit_transform(x_treino) #normalizando x_treino
x_teste_normalizado = norm.fit_transform(x_teste) #normalizando x_teste
x_val_normalizado = norm.fit_transform(x_val) #normalizando x_val
```

```python
#Importando modelo
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier() #instanciando modelo
knn.fit(x_treino_normalizado, y_treino) #treinando modelo


knn_score_teste = knn.score(x_teste_normalizado,y_teste) #score do modelo com dados de teste
knn_score_val = knn.score(x_val_normalizado,y_val) #score do modelo com dados de validação

print(f'A taxa de acerto do modelo KNeighborsClassifier para os dados de teste é de {knn_score_teste:.2%}')
print(f'A taxa de acerto do modelo KNeighborsClassifier para os dados de validação é de {knn_score_val:.2%}')
```

```python
#Prevendo resultados para o x_teste e x_val
knn_predict_teste = knn.predict(x_teste_normalizado)
knn_predict_val = knn.predict(x_val_normalizado)
```

```python
#Matriz de confusão DecisionTreeClassifier para dados de teste

print(f'Acuracia Teste: {accuracy_score(y_teste,knn_predict_teste):.2%}')
print(f'Precisão Teste: {precision_score(y_teste,knn_predict_teste):.2%}')
print(f'Recall Teste: {recall_score(y_teste,knn_predict_teste):.2%}')
print(f'F1 Score Teste: {f1_score(y_teste,knn_predict_teste):.2%}')
```

```python
#Matriz de confusão DecisionTreeClassifier para dados de validação

print(f'Acuracia Teste: {accuracy_score(y_val,knn_predict_val):.2%}')
print(f'Precisão Teste: {precision_score(y_val,knn_predict_val):.2%}')
print(f'Recall Teste: {recall_score(y_val,knn_predict_val):.2%}')
print(f'F1 Score Teste: {f1_score(y_val,knn_predict_val):.2%}')
```

```python
#Matriz de confusão dos dados de treino para o KNN

vizualizacaoknn = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_teste,knn_predict_teste))
vizualizacaoknn.plot()

plt.show()
```
